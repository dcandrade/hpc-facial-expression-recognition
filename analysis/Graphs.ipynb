```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('../results/output_1000epochs_8093train_2013test___2950098091.txt')
df.set_index('epoch', inplace=True)
```

```python
print(len(df))
df.head()
```

```python
test = df.loc[-1]
train = df.loc[1:1000]
test
```

```python
plot = train.accuracy.plot()
plot.set_title("Acur√°cia por √âpoca")
plot.set_xlabel("# √âpoca")
plot.set_ylabel("Acur√°cia")
```

```python
prplot = train[['precision', 'recall']].set_index('recall').plot()
prplot.set_title("Curva Precision-Recall")
prplot.set_xlabel("Recall")
prplot.set_ylabel("Precision")
```

```python
costplot = train['cost'].plot()
costplot.set_title("Custo por √âpoca de Treinamento")
costplot.set_xlabel("√âpoca")
costplot.set_ylabel("Custo")
```

```python
costplot = train['f1'].plot()
costplot.set_title("F1 Score por √âpoca de Treinamento")
costplot.set_xlabel("√âpoca")
costplot.set_ylabel("F1 Score")
```
